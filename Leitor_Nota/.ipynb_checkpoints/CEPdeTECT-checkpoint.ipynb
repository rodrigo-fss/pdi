```python
import cv2
import numpy as np
```

```python
#
class Blob:
    def __init__(self, x, y, area, peri, rect, corners):
        #Coordinates of centroid, area and perimeter
        #calculated by cv2.moments()
        self.x = x
        self.y = y
        self.area = area
        self.peri = peri
        
        #cv2.minAreaRect(). Contains following detals:
        #( center (x,y), (width, height), angle of rotation )
        self.rect = rect
        
        # cv2.boxPoints(). Contains all four coordinates
        self.corners = corners
```

```python
#Recebe uma imagem em GRAYSCALE e procura por
#blobs em sequência horizontalmente
def detecta_cep(img, rotacao_max, area_min, area_max, deslocamento_max_x):
    ret, imgthresh = cv2.threshold(img,170,255,cv2.THRESH_BINARY_INV)
    cv2.imwrite('im2_bin.jpg', imgthresh)
        
    #RETR_EXTERNAL: seleciona contornos externos
    #CHAIN_APPROX_TC89_KCOS: Teh-Chin chain, método de aproximação de contorno
    #Determina a região de suporte pra cada ponto,
    #computa curvatura para cada,
    #determina ponto dominante por processo de supressão dos não máximos
    contorno = cv2.findContours(imgthresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_TC89_KCOS) 
    
    img_black = np.zeros(img.shape[:2], np.uint8)
    img_quad = cv2.imread("1.jpg", cv2.IMREAD_COLOR)

    stack = []
    last_blob = Blob(0,0,0,0, [], (0,0))
    actual_blob = Blob(0,0,0,0, [], (0,0))
    
    #Itera sobre todos contornos
    for c in contorno[0]:
        #Calcula perímetro e identifica formato
        actual_blob.peri = cv2.arcLength(c,True)
        
        #Ramer–Douglas–Peucker algorithm, simplifica traços
        approx = cv2.approxPolyDP(c, 0.116*actual_blob.peri, True)
        
        #Se tem aproximadamente quatro vértices
        if len(approx) == 4:
            #Desenha todos contornos quadriláteros
            cv2.drawContours(img_black, [c], -1, (255,255,255), 2)
            
            #Moments da imagem, auxilia em vários cálculos, média ponderada dos pixels
            M = cv2.moments(c)    
            actual_blob.x = int(M['m10']/M['m00'])
            actual_blob.y = int(M['m01']/M['m00'])
            actual_blob.area = cv2.contourArea(c)
            actual_blob.rect = cv2.minAreaRect(c)
            actual_blob.corners = np.int0(cv2.cv.BoxPoints(actual_blob.rect))
                    
            #Candidato a CEP
            if -rotacao_max < actual_blob.rect[2] < rotacao_max \
            and area_min < actual_blob.area < area_max:
            #and 55 < actual_blob.peri < 350 \
                
                #Analisa quantos pixels o blob se deslocou horizontalmente
                if actual_blob.x < last_blob.x + deslocamento_max_x:
                    #desenha e põe na pilha
                    cv2.drawContours(img_quad,[actual_blob.corners],0,(0,0,255),2)
                    #(self, x, y, area, peri):
                    stack.append(Blob(actual_blob.x,actual_blob.y, \
                                      actual_blob.area,actual_blob.peri, \
                                      actual_blob.rect, actual_blob.corners))              
                elif len(stack) > 1:
                    del stack[:]
        #end if
        last_blob = actual_blob
    #end for
    
    cv2.imwrite('im3_contornos.jpg', img_black)
    cv2.imwrite('im4_quadrados.jpg', img_quad)
    
    return stack
```

```python
img1 = cv2.imread("1.jpg", cv2.IMREAD_GRAYSCALE )
cv2.imwrite('im1_gray.jpg', img1)
#img1 = cv2.GaussianBlur(img1,(3,3),1)

#Img-GRAYSCALE, MaxRotation, MinArea, MaxArea, Max-X-Jump
stack = detecta_cep(img1, 2, 1400, 3000, 60)

for s in stack:
    print "x,y of centroid: (" + str(s.x) + "," + str(s.y) + \
    ")\narea: " + str(s.area) + "\nperimeter: "  + \
    str(s.peri) + "\nwidth: " + str(s.rect[1][0]) + \
    "\nheight: "+ str(s.rect[1][1]) + "\ncorners:"
    #rect: ( center (x,y), (width, height), angle of rotation )
    for c in s.corners:
        print str(c),
    print '\n'
```
